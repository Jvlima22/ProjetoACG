### Exibindo Menu Principal com 4 opções:
- 1° Login;
- 2° Cadastro;
- 3° Cálculos;
- 4° Sair.

### Exibindo Menu de Cálculos com 5 opções:
- 1° Cálculo de Confirmação de Metragem;
- 2° Cálculo de Peso e Metragem;
- 3° Cálculo de Peso e Tempo;
- 4° Cálculo de Diâmetro e Peso;
- 5° Voltar.

### Declarando as funções do Menu Principal:
- Login: Usuário e Senha;
- Cadastro: Nome, Usuário e Senha;
- Cálculos: CCM, CPM, CPT e CDP
- Sair: Encerrar o programa e exibir mensagem.

### Declarando as funções do Menu de Cálculos:
- 1. Cálculo de Confirmação de Metragem (CCM): n1, pesoEspecifico, espessura, largura, metragem, nome e setor
- 2. Cálculo de Peso e Metragem (CPM): n1, espessura, largura, tempoRestante, pesoEspecifico, velocidade, nome e setor 
- 3. Cálculo de Peso e Tempo (CPT):  n1, largura, PesoMax, pesoMedio, diametroMax, pesoAprox, nome e setor
- 4. Cálculo de Diâmetro e Peso (CDP): n1, pesoEspecifico, espessura, largura, puxador, tempoRestante, nome e setor
- 5. Voltar: Retorna ao Menu Principal.



## Autenticar acesso desse programa 
- Cadastro do usuário
- Logar caso usuário tenha cadastro
- Caso ele não tenha deverá se cadastrar 
- Redefinir senha 

```python
def menu_principal():
    while True:
        print("\nMenu Principal:")
        print("1. Login")
        print("2. Cadastro")
        print("3. Cálculos")
        print("4. Sair")

        escolha = input("Escolha uma opção: ")

        if escolha == "1":
            # Funcionalidade de Login
            print("\nFaça login para acessar o sistema!\n")

            usuarios = input("Digite seu usuário: ")
            senha = input("Digite sua senha: ")

            if usuarios == "jvlima22_" and senha == "josu190602":
                print("\nLogin efetuado com sucesso!\n")
            else:
                print("\nUsuário ou senha inválidos!\n")

        elif escolha == "2":
            # Funcionalidade de Cadastro
            print("\nCadastre-se para acessar o sistema!\n")

            nome = input("Digite seu nome: ")
            usuarios = input("Digite seu usuário: ")
            senha = input("Digite sua senha: ")

            print("\nCadastro efetuado com sucesso!\n")

        elif escolha == "3":
            # Menu de cálculos
            menu_calculos()

        elif escolha == "4":
            print("Saindo do programa. Até logo!")
            break

        else:
            print("Opção inválida. Tente novamente.")

def menu_calculos():
    while True:
        print("\nMenu de Cálculos:")
        print("1. Cálculo de Confirmação de Metragem")
        print("2. Cálculo de Peso e Metragem")
        print("3. Cálculo de Peso e Tempo")
        print("4. Cálculo de Diâmetro e Peso")
        print("5. Voltar")

        escolha = input("Escolha uma opção de cálculo: ")

        if escolha == "1":
            # Funcionalidade do cálculo de Confirmação de Metragem
            print("\nCálculo de Confirmação de Metragem\n")

            n1 = int(input("Digite o Número da Bobina: "))
            pesoEspecifico = float(input("Digite a P.E da Ordem de Serviço: "))
            espessura = float(input("Digite a Espessura da Bobina: "))
            largura = float(input("Digite a Largura da Bobina: "))
            metragem = float(input("Digite a Metragem Total da Bobina: "))
            nome = input("Informe o Nome do Funcionário: ")
            setor = input("Informe o Setor da Empresa: ")

            pesoTotal = (pesoEspecifico * espessura * largura * metragem)

            print("\nNome do Funcionário:", nome)
            print("Setor da Empresa:", setor)
            print("Número da Bobina:", n1)
            print("P.E da Bobina: %.2f" % pesoEspecifico)
            print("Espessura da Bobina:", espessura)
            print("Largura da bobina:", largura)
            
            resultado_arredondado = round(pesoTotal, 3)
            pesoTotal_str = str(resultado_arredondado)[:3]
            print("Peso Total da Bobina:", pesoTotal_str)

        elif escolha == "2":
            # Funcionalidade do cálculo de Peso e Metragem
            print("\nCálculo de Peso e Metragem!")

            n1 = int(input("Informe o Número da Bobina: "))
            espessura = float(input("Informe a Espessura da Bobina: "))
            largura = float(input("Informe a Largura da Bobina: "))
            tempoRestante = float(input("Informe o Tempo Restante para o término da Bobina: "))

            if tempoRestante <= 60:
                print("\nO Tempo Restante está dentro dos parâmetros!\n")
            else:
                print("\nO Tempo Restante não está dentro dos parâmetros!\n")
                break

            pesoEspecifico = float(input("Informe a P.E da Ordem de Serviço: "))
            velocidade = float(input("Informe a Velocidade da Máquina: "))
            nome = input("Informe o Nome do Funcionário: ")
            setor = input("Informe o Setor da Empresa: ")

            metragem = tempoRestante * velocidade
            pesoTotal = metragem * pesoEspecifico * largura * espessura

            print("\nNome do Funcionário:", nome)
            print("Setor da Empresa:", setor)
            print("Número da Bobina:", n1)
            print("Largura da Bobina:", largura)
            print("Espessura da Bobina:", espessura)
            print("Velocidade da Máquina:", velocidade)
            print("Tempo Restante foi de:", tempoRestante)
            
            resultado_arredondado = round(pesoTotal, 3)
            pesoTotal_str = str(resultado_arredondado)[:3] 
            print("O Peso Total dessa Bobina foi de:", pesoTotal_str)

        elif escolha == "3":
            # Funcionalidade do cálculo de Peso e Tempo
            print("\nCálculo de Diâmetro e Peso!\n")

            n1 = int(input("Informe o Número da Bobina: "))
            largura = float(input("Digite a Largura da Bobina: "))
            pesoMax = float(input("Digite o Peso Máximo da Bobina: "))
            nome = input("Informe o Nome do Funcionário: ")
            setor = input("Informe o Setor da Empresa: ")

            pesoMedio = ((pesoMax + (9.828434 * largura)) / (1044 * largura))
            diametroMax = (pesoMedio ** 0.5) * 1000  # Usamos ** 0.5 para calcular a raiz quadrada
            pesoAprox = 1350 * largura * (0.785398 * diametroMax * diametroMax - 0.0073898)

            print("\nNome do Funcionário:", nome)
            print("Setor da Empresa:", setor)
            print("Número da Bobina:", n1)
            print("Largura da Bobina:", largura)
            print("Peso Máximo da Bobina:", pesoMax)
            
            resultado_arredondado = round(pesoAprox, 3)
            pesoAprox_str = str(resultado_arredondado)[:3] 
            print("Peso Aproximado da Bobina:", pesoAprox_str)

        elif escolha == "4":
            # Funcionalidade do cálculo de Diâmetro e Peso
            
            print("\nCálculo de Peso e Tempo!\n")

            n1 = int(input("Informe o Número da Bobina: "))
            pesoEspecifico = float(input("Informe a P.E da Ordem de Serviço: "))
            espessura = float(input("Informe a Espessura da Bobina: "))
            largura = float(input("Informe a Largura da Bobina: "))
            puxador = float(input("Informe o Puxador da Máquina: "))
            tempoRestante = float(input("Informe o Tempo Restante para o término da Bobina: "))
            nome = input("Informe o Nome do Funcionário: ")
            setor = input("Informe o Setor da Empresa: ")

            metragem = tempoRestante * puxador
            pesoTotal = metragem * pesoEspecifico * largura * espessura

            print("\nNome do Funcionário:", nome)
            print("Setor da Empresa:", setor)
            print("Número da Bobina:", n1)
            print("Espessura da Bobina:", espessura)
            print("Largura da Bobina:", largura)
            print("Tempo Restante da Bobina:", tempoRestante)
            
            resultado_arredondado = round(pesoTotal, 3)
            pesoTotal_str = str(resultado_arredondado)[:3]
            print("Peso Total da Bobina:", pesoTotal_str)

        elif escolha == "5":
            # Voltar para o menu principal
            break

        else:
            print("Opção inválida. Tente novamente.")

# Inicie o programa chamando o menu principal
menu_principal()
```
